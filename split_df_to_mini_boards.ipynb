```python
from split_mini_boards_functions import *
import pickle
```

```python
SIZE = 4
AMOUNT_BOARDS = 1000
AMOUNT_MOVES = 100
NUM_DICT = 1
SPLIT_SIZE = 3
```

```python
gen = 2
name_df = f'{PATH_DF}\\{SIZE}-{AMOUNT_BOARDS}\\{SIZE}size_{AMOUNT_BOARDS}boards_{gen}gen_reverse'
reverse_df = pd.read_pickle(f'{name_df}.pkl')
reverse_df
```

```python
new_columns = [f'Col_{i}' for i in range(SPLIT_SIZE*SPLIT_SIZE*gen)]
dfs = []
for pixel in range(SIZE*SIZE):
    res_df = pd.DataFrame(columns=new_columns)
    for i, row in reverse_df.iterrows():
        print_big_numbers(i)
        selected_pixels = generate_split_indexes_numbers(SIZE,SPLIT_SIZE,pixel)
        new_row = extract_selected_columns(SIZE, row, selected_pixels)
        res_df = pd.concat([res_df,new_row])
    dfs.append(res_df)
```

```python
size = SIZE
amount_boards = AMOUNT_BOARDS
reverse = True
path = PATH_DF
for i in range(len(dfs)):
    os.makedirs(f'{path}\\{size}-{amount_boards}-mini-boards-to-{SPLIT_SIZE}',exist_ok=True)
    if reverse:
        name_df = f'{path}\\{size}-{amount_boards}-mini-boards-to-{SPLIT_SIZE}\\pixel{i}_{size}size_to{SPLIT_SIZE}_{amount_boards}boards_{gen}gen_reverse'
    else:
        name_df = f'{path}\\{size}-{amount_boards}-mini-boards-to-{SPLIT_SIZE}\\pixel{i}_{size}size_to{SPLIT_SIZE}_{amount_boards}boards_{gen}gen_non-reverse'
    dfs[i].to_pickle(f'{name_df}.pkl')
```
