```python
def generate_split_indexes_numbers(size,split_size,pixel):
    i_pixel,j_pixel = find_loc(pixel,size)
    #print(i_pixel,j_pixel)
    n = split_size-2
    result = []
    for i in range(-n, n+1):
        for j in range(-n, n+1):
            #print(i,j)
            
            i_res = i+i_pixel
            if i_res <0:
                i_res += size
            if i_res >= size:
                i_res -= size
            
            j_res = j+j_pixel
            if j_res <0:
                j_res += size
            if j_res >=size:
                j_res -= size
            
            result.append(find_pixel(i_res, j_res,size))
    return result
```

```python
def extract_selected_columns(size, row, selected_cols):
    num_cols = len(row)
    num_blocks = num_cols // (size*size)
    selected_data = []

    for block_idx in range(num_blocks):
        offset = block_idx * (size*size)
        for col in selected_cols:
            col_name = f'Col_{offset + col + 1}'
            selected_data.append(row[col_name])

    return pd.DataFrame([selected_data], columns=[f'Col_{i+1}' for i in range(len(selected_data))])
```

```python
new_columns = [f'Col_{i}' for i in range(1, SPLIT_SIZE*SPLIT_SIZE*gen)]
dfs = []
for pixel in range(SIZE*SIZE):
    res_df = pd.DataFrame(columns=new_columns)
    for i, row in reverse_df.iterrows():
        print_big_numbers(i)
        selected_pixels = generate_split_indexes_numbers(SIZE,SPLIT_SIZE,pixel)
        new_row = extract_selected_columns(SIZE, row, selected_pixels)
        res_df = pd.concat([res_df,new_row])
    dfs.append(res_df)
```
